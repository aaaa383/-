```python
#データ結合
combined = pd.concat([train,test],axis=0)
print(combined.shape)
combined.head()
```

```python
#前処理
combined["月"] = combined["gameday"].apply(lambda x :x.split("/")[0])
combined['月'] = combined['月'].astype(int)

combined['humidity'] = combined['humidity'].str.replace('%', '')
combined['humidity'] = combined['humidity'].astype(int)
```

```python
#曜日
combined["月"] = combined['gameday'].map(lambda x: 1 if '月' in str(x) else 0)
combined["火"] = combined['gameday'].map(lambda x: 1 if '火' in str(x) else 0)
combined["水"] = combined['gameday'].map(lambda x: 1 if '水' in str(x) else 0)
combined["木"] = combined['gameday'].map(lambda x: 1 if '木' in str(x) else 0)
combined["金"] = combined['gameday'].map(lambda x: 1 if '金' in str(x) else 0)
combined["土"] = combined['gameday'].map(lambda x: 1 if '土' in str(x) else 0)
combined["日"] = combined['gameday'].map(lambda x: 1 if '日' in str(x) else 0)
combined["祝"] = combined['gameday'].map(lambda x: 1 if '祝' in str(x) else 0)

#テレビ局
combined["スカパー！"] = combined['tv'].map(lambda x: 1 if 'スカパー！' in str(x) else 0)
combined["スカパー！プレミアム"] = combined['tv'].map(lambda x: 1 if 'スカパー！プレミアムサービス' in str(x) else 0)
combined["NHK"] = combined['tv'].map(lambda x: 1 if 'ＮＨＫ' in str(x) else 0)
combined["e2"] = combined['tv'].map(lambda x: 1 if 'ｅ２' in str(x) else 0)
```

```python
del combined['tv']
del combined['gameday']
```

```python
#試合開始時間
combined["time"] = combined["time"].apply(lambda x :x.split(":")[0])
combined['time'] = combined['time'].astype(int)

#第何節か
combined["match_"] = combined["match"].apply(lambda x :x.split("節")[0])
combined['match_'] = combined['match_'].str.replace('第', '')
combined['match_'] = combined['match_'].astype(int)
```

```python
#何日目か
combined["match_day"] = combined["match"].apply(lambda x :x.split("節")[1])
combined['match_day'] = combined['match_day'].str.replace('第', '').str.replace('日', '')
combined['match_day'] = combined['match_day'].astype(int)
del combined['match']
```

```python
#曜日
combined["晴"] = combined['weather'].map(lambda x: 1 if '晴' in str(x) else 0)
combined["曇"] = combined['weather'].map(lambda x: 1 if '曇' in str(x) else 0)
combined["雨"] = combined['weather'].map(lambda x: 1 if '雨' in str(x) else 0)
combined["屋内"] = combined['weather'].map(lambda x: 1 if '屋内' in str(x) else 0)
combined["のち"] = combined['weather'].map(lambda x: 1 if 'のち' in str(x) else 0)
combined["一時"] = combined['weather'].map(lambda x: 1 if '一時' in str(x) else 0)
combined["時々"] = combined['weather'].map(lambda x: 1 if '時々' in str(x) else 0)
```

```python
#使うデータ
combined2 = combined[['away_team','capa','home_team','humidity','referee','stadium','stage','temperature','time','y','year','月',
                     '火','水','木','金','土','日','祝','スカパー！','スカパー！プレミアム','NHK','e2','match_','match_day','晴',
                     '曇','雨','屋内','のち','一時','時々']]
```

```python
#ダミー変数化
combined = pd.get_dummies(combined, drop_first=True) # drop_first=Trueでk-1個のダミー変数となる
```

## モデル作成

```python
#trainとtestに分割
train_x = combined.head(1721)
test_x = combined.tail(313)
train_y = train_x['y']

del train_x['y']
del test_x['y'] 
```

```python
print(train_x.shape)
print(test_x.shape)
```

```python
from tqdm import tqdm
from lightgbm import LGBMRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import KFold, train_test_split

kf = KFold(n_splits = 15,
           shuffle = True,
           random_state=1)

preds15 = pd.DataFrame()

for i, (tr_idx, va_idx) in tqdm(enumerate(kf.split(train_x))):
    
    trtr_x, trva_x = train_x.iloc[tr_idx].values, train_x.iloc[va_idx].values
    trtr_y, trva_y = train_y.iloc[tr_idx].values, train_y.iloc[va_idx].values
    
    model = LGBMRegressor(random_state=1,
                          min_child_samples = 60,
                          n_estimators = 200,
                          num_leaves=500)
    
    model.fit(trtr_x,trtr_y)
    pred_test = model.predict(test_x.values)
    preds15[i] = pred_test
    
EnsemblePred = preds15.mean(axis=1)

submit = test[["id"]].copy()
submit["pred"] = EnsemblePred
```

```python
# Train
submit.to_csv(
    "../submit/submit_LGBM4_20201106_1.csv",  sep=",", index=False, header=False)
```
